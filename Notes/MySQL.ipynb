# Database Keys :
A Database key is an attribute or a set of attributes which helps us uniquely identify a row in a table.

## Types of Keys :
   1. Super Key
   2. Candidate Key
   3. Primary Key
   4. Alternate Key
   5. Foreign Key
   6. Composite Key
   7. Compound Key
   8. Surrogate Key

### Super Key :

Let's assume a dataset as follows :

|EmpID|Aadhar Card|Emp.Name|DOB|Salary|Email|
|-|-|-|-|-|-|

Now in this, the columns having unique values would be {EmpID,Aadhar,Email}

A **Super Key** is a superset of all these unique columns :
   * {EmpID,Aadhar,Email, EmpID+Aadhar, EmpID+Email, Aadhar+Email, EmpID+Aadhar+Email}

### Candidate Key:
It is the minimum number of columns required to become a key.

from Super keys:
   * {EmpID,Aadhar,Email, EmpID+Aadhar, EmpID+Email, Aadhar+Email, EmpID+Aadhar+Email}
   * {EmpID,Aadhar,Email} would be the **Candidate Keys** 
   * {EmpID+Aadhar, EmpID+Email, Aadhar+Email, EmpID+Aadhar+Email} would be rejected as with just 1 column we could make a key so why have 2?

### Primary Key :
It is that **Candidate key** which is finally selected to be the one that identifies the rows in a table uniquely or differentiate between 2 rows.

Let's say we select EmpID as the **Primary Key**

##### Requirements for a column to be a **Primary Key**
   1. It should have unique values.
   2. It should not have Null Values.
   
##### Good to have :
   1. Preferably Numerical.
   2. Should be as small as possible.

### Alternate Key :
**Alternate keys** are those keys which are rejected to become a **Primary Key** from the set of **Candidate Keys**.

In our case, they would be : {Aadhar , Email}

### Foreign Key:
A **Foreign Key** is a key that is a **Primary Key** in one or more other Tables.

Let's assume 2 Tables as below :

| Branch ID| Branch Name | HOD |
|-|-|-|

and another Table :

|Student ID|Name|Branch ID|
|-|-|-|

Here, Branch ID would be a **Primary Key** in the 1st Table and a **Foreign Key** in the 2nd Table.

### Composite Keys :

Combination of 2 keys is called as a **Composite Key**.

Eg. {EmpID+Email}, {Email+Aadhar}

### Compound Keys :
**Compound Key** is similar to **Composite key** but in this atleast 1  key is a **Foreign Key**.

### Surrogate Keys :
These are those Keys which are created, when we do not have any columns that can be considered as keys in a table.

Eg :

|Name|DOB|City|
|-|-|-|

in the table above, we cannot consider any column as a key hence, we introduce another column ID as shown below. Which acts as a key.

|ID|Name|DOB|City|
|-|-|-|-|

#### Constraints in SQL :

##### NOT NULL :
When NOT NULL is specified on a column, it cannot be left empty. It needs to have some info or else SQL will fill in a value as 0 for integer and Decimal data types but for a Varchar, text i'd leave it as blank.

Eg:
   - CREATE TABLE Users(

##### UNIQUE :
It makes sure that all the records entered into the column with this constraint are unique or else, it throws up an error as shown below:

    CREATE TABLE Users(
        ID integer UNIQUE NOT NULL,
        Name varchar(255),
        Email varchar(255),
        Phone integer UNIQUE
    )



In the above example we tried to put in a constrain that the values in the ID column should not be null and they should be unique but while inserting the records/values we left the values blank for ID column. As we left few of the values, blank SQL inserted the values as 0 to all the Null values but then there was unique constraint added according to which 2 values cannot be same. Hence the above error.

- Other way of writing the command
    CREATE TABLE Users(
        ID integer UNIQUE NOT NULL,
        Name varchar(255),
        Email varchar(255),
        Phone integer,
        UNIQUE(Name, Email)
    )
 
With the above command, the table would not allow the insertion of a data where the combination of Name and Email is the same for 2 Records. This works for varchars only.



    CREATE TABLE Users(
        ID integer UNIQUE NOT NULL,
        Name varchar(255),
        Email varchar(255),
        Phone integer,
        CONSTRANT User_Email UNIQUE(Email)
    )
The above command would allow us to make changes to Constrant User_Email at a later stage if needed.

### Types of **JOINS** :
   1. Cartesian Product // Cross Join
   2. Inner Join
   3. Self Join
   4. Outer Join
       - Left Outer
       - Right Outer
       - Full Outer

#### CROSS JOIN // Cartesian Product :
This would provide a Cartesian product or every possible combination. Eg

Lets assume we have 2 tables, Table 1 has 3 rows A,B,C and Table 2 has 2 rows X,Y.

The cartesian product would yield us 6 combinations {(A,X),(A,Y),(B,X),(B,Y),(C,X),(C,Y)} this type of JOIN is not too useful but is a base for other types of joins.





#### Inner JOIN :
INNER JOIN keyword selects records that have matching values in both tables.

    SELECT Table A.Student ID, Table A.Name, Table B.Department 
    FROM Table A 
    INNER JOIN Table B
    ON Table A.Student ID = Table B.Student ID





#### LEFT JOIN // LEFT Outer JOIN :
Shows all the rows for matching columns from Table A and Table B + all the unmatched rows from Table A (Table on the Left Side)

    SELECT table_a.StudentID,table_a.Name,table_b.Department 
    FROM table_a LEFT JOIN table_b
    ON table_a.StudentID=table_b.StudentID



#### RIGHT JOIN // RIGHT Outer JOIN :
Here, all the rows for matching column + all the unmatched rows from Table b (Right Table) are displayed.

    SELECT table_a.StudentID,table_a.Name,table_b.Department 
    FROM table_a RIGHT JOIN table_b
    ON table_a.StudentID=table_b.StudentID



##### UNION 
The UNION operator is used to combine the result-set of two or more SELECT statements.

*Every SELECT statement within UNION must have the same number of columns
*The columns must also have similar data types
*The columns in every SELECT statement must also be in the same order. 

    SELECT StudentID FROM table_a
    UNION
    SELECT StudentID FROM table_b



If we want to see all the records and not just the Unique, we use UNION ALL

    SELECT StudentID FROM table_a 
    UNION ALL 
    SELECT StudentID FROM table_b;



#### FULL OUTER JOIN :
There is no direct way for full outer join in SQL but all the websites do show there is which does not work for us. Hence we run query for LEFT JOIN UNION query for RIGHT JOIN.

    SELECT *
    FROM table_a LEFT JOIN table_b
    ON table_a.StudentID = table_b.StudentID
    
    UNION
    
    SELECT *
    FROM table_a RIGHT JOIN table_b
    ON table_a.StudentID = table_b.StudentID

### Sub-Query :
Sub-Queries are Nested Queries (One Query inside another).These are of 2 types:
   1. Independent
   2. Correlated 

#### Independent Sub-Query :
These are those queries where the inner query is Independent of the outer Query or Inner Query has nothing to do with the outer Query.

Eg:

**Finding the details of the movie with Highest Budget:**

    Select MovieName,LeadStar,Director,Revenue FROM data_for_repository 
    WHERE Revenue = (SELECT MAX(Revenue) FROM `data_for_repository`)
    


Here the Inner Query can run Independently or as a separate Query and has nothing to do with the outer Query.

Note : Max is an Aggrigate Function and needs only the column on which the operation has to be performed.
