#### Types of Sampling Techniques :
   1. Simple Random Sampling : 
    Simply picking up random data/people. Here every member of population (N) has equal chance of getting picked as a sample (n).
   
   2. Stratified Sampling :
    It is a technique where in the population is split into non-overlapping groups and then the samples are picked. Eg. A survey based on gender, where the population(N) is divided in to 2 and then the samples are picked.
    
   Eg. Where Stratified sampling would not work ---> Sampling the students in the class based on their hobbies (A student can have multiple Hobbies)
   
  
  3. Systematic Sampling : 
   In this type, we pick up every nth individual/ sample from population (N).
   
   Eg. In a manufacturing company, we pick every 10th sample to check the quality from a batch.
   
  4. Convenience Sampling :
       Eg. we are doing a survey on Photography and only those people who are interested in Photography are considered as a sample. Domain based sampling.

#### Z-Score :

Let's say we have a distribution with mean(µ) = 4 and standard deviation(σ) = 1 and we want to represent a point 4.5 it'd be easy to plot and say 4.5 lies .5 standard deviation from the mean.

But assume a scenario where we have mean as 4.32475, standard deviation is 1.349 and we want to plot 4.5 it's be difficult for us to say how many standard deviation does this point lie from the mean. Hence the concept of **Z-Score**.

**Z-Score** helps us understand how far is a particular point away from the mean.



now assume the same scenario where we have mean as 4.32475, standard deviation is 1.349 and we want see how far is 4.5, from the mean.

Z-Score (z) = (4.5 - 4.32475)/1.349
Z-Score (z) = (0.17525000000000013)/1.349
Z-Score (z) = 0.13

which says the point 4.5 lies 0.13 standard deviation towards right from the mean.

### Addition Rule(Probability, "or") :

#### Mutual Exclusive Events :
2 events are mutually exclusive if they cannot occur at the same time.

Eg:
   -  In a six-sided die, the events “2” and “5” are mutually exclusive. We cannot get both the events 2 and 5 at the same time when we threw one die.
   
       - What is the probability of a die showing a number 3 or number 5?

       - Solution: Let,

        P(3) is the probability of getting a number 3

        P(5) is the probability of getting a number 5

        P(3) = 1/6 and P(5) = 1/6

        So,

        P(3 or 5) = P(3) + P(5)

        P(3 or 5) = (1/6) + (1/6) = 2/6

        P(3 or 5) = 1/3

Therefore, the probability of a die showing 3 or 5 is 1/3.


   - When tossing a coin, the event of getting head and tail are mutually exclusive. Because the probability of getting head and tail simultaneously is 0.
   
   
   - A card is drawn at random from a well-shuffled deck of 52 cards. Find the probability that the card drawn is a king or an ace.

       - Answer: 

        As per the definition of mutually exclusive events, selecting an ace and selecting a king from a well-shuffled deck of 52 cards are termed mutually exclusive events.

        Assume X to be the event of drawing a king and Y to be the event of drawing an ace.

        In a standard deck of 52 cards, there exists 4 kings and 4 aces.

        P (an event) = count of favourable outcomes / total count of outcomes

        P (selecting a king from a standard deck of 52 cards) = P (X) = 4 / 52 = 1 / 13

        P (selecting an ace from a standard deck of 52 cards) = P (Y) = 4 / 52 = 1 / 13

        To compute P (king or ace).

        By the formula of addition theorem for mutually exclusive events,

        P (X U Y) = P (X) + P (Y)

        P (X U Y) = (1 / 13) + (1 / 13)

        = (1 + 1) / 13

        = 2 / 13

        The probability of selecting a king or an ace from a well-shuffled deck of 52 cards = 2 / 13.

## Permutation and Combination

### Permutation :

The term permutation refers to a mathematical calculation of the number of ways a particular set can be arranged. Put simply, a permutation is a word that describes the number of ways things can be ordered or arranged. With permutations, the order of the arrangement matters. 

##### Permutation Formula :
If the total number of data is “n” and the choice is of “r” things, then permutation will be (without replacement and regarding an order)-

**$_nP_r$ = (n!) / (n-r)!**

Eg:

   - A company wants to build out its warehouse network across the country. The company will commit to two locations out of five possible sites. Order matters because they will be built sequentially. The number of permutations is:

     P(5,2) = 5! ÷ (5-2)! = 5! ÷ 3! = 20
     
**It is basically the number of ways a combination of 2 Cities can occur out of 5 eg: Let Bangalore, Hyderabad, Pune, Chennai, Noida be the 5 selected destinations to build an office. Now the company wants to build 2 offices in 2 cities out of these identified cities**

### Combination :
Combination is again the arrangement of 2 cities out of 5 like the above example but here a combination cannot be repeated by interchanging the values.

##### Combination Formula :
From a group of “n” data, the selection of “r” things without regarding order and replacement-

**$_nC_r$ = $_nP_r$ / r!** 

**$_nC_r$ = n! / r!(n-r)!** 

Moreover, the relation between these two is $_nC_r$ = $_nP_r$ / r!.

Eg:

     C(5,2) = 5! ÷ (5-2)! = 5! ÷ (2! x 3!)

## Hypothesis Testing, Confidence Interval, Significance Value :

### Hypothesis Testing :
Hypothesis testing is an act in statistics whereby an analyst tests an assumption regarding a population parameter.

Eg: Proving a coin is fair by tossing it 100 times. For it to be Fair, we should ideally get heads 50 times and tails 50 times.

##### Steps Involved in Hypothesis Testing :
   1. Determine the Null Hypothesis (H0) : Coin is Fair
   2. Determine Alternate Hypothesis (H1): Coin is not Fair.
   3. Experiment
   4. Reject or Accept Null Hypothesis.
   
### Significance Value(α) : 
Assume a situation where in we toss a coin 100 times and get heads 40 times, do we call it fair or unfair?

That's were the concept of Significance Value comes into the picture.

It is a range with in which if the total number of times head or tail occurs it would be termed as Fair or in other words how far away the result of the coin toss experiment be from the mean and still be termed as fair.

α is determined by the domain experts it is usually 0.05.

### Confidence Interval :
(1-Significance Value)*100

##### Point Estimate : 
The value of any statistic that estimates the value of a parameter.

Eg:

Estimating the value of Population Mean(μ) with the help of sample mean(x̅). Here x̅ would be approximately equal to Population mean(μ). Sample mean(x) may be higher or may be lower than μ.

**Confidence Intervals are defined so that we could come towards the mean**

#### Finding the Interval or the Range  :
   * Formula : Point Estimate +/- Margin of Error

## Type 1 and Type 2 Error :

Assume a situation where a snake bites a person and we need to take a decision if the snake is poisonous or no and if the patient needs to be taken to the hospital or no.

   - Null Hypothesis : Snake is not poisonous and the patient should not be taken to the hospital. (Is innocent until proven guilty)
   - Alternate Hypothesis : Snake is poisonous and the patient should be taken to the hospital.
   
Now we need to take a decision, if Null Hypothesis is true and it should be accepted or Null Hypothesis is false and it should be rejected.

**Outcomes of the Decision :**
   - Outcome 1 : Null Hypothesis is accepted and in reality is True. (We accepted the Snake is not poisonous and it was not poisonous.
   
   - Outcome 2: Null Hypothesis is rejected when in reality it is True. (We assumed the snake to be poisonous when it was not and took the patient to the doctor). **Type-1 Error**.
   
   - Outcome 3: Null Hypothesis is rejected when in reality it is not true or in other words Alternate Hypothesis is accepted when in reality it is true. (Snake that bit the person was poisonous and we take the patient to the hospital.
   
   - Outcome 4: Null Hypothesis is accepted when in reality it is not true or in other words, Alternate Hypothesis is rejected when it is true.(Snake that bit the person was considered to be non-poisonous and the patient was not taken to the hospital, when in reality it was poisonous and the patient should have been taken down to the hospital).**Type-2 Error**
   


## One-tailed & Two-Tailed Test :

### One-tailed Test : 
One-tailed tests allow for the possibility of an effect in one direction.

**One-tailed test: Alternative hypothesis contains either < or > sign**


Eg:

Suppose it’s assumed that the average weight of a certain widget produced at a factory is 20 grams. However, one engineer believes that a new method produces widgets that weigh less than 20 grams.

   - H0 (Null Hypothesis): μ ≥ 20 grams
   - HA (Alternative Hypothesis): μ < 20 grams
   
**Note**: We can tell this is a one-tailed test because the alternative hypothesis contains the less than (<) sign. Specifically, we would call this a left-tailed test because we’re testing if some population parameter is less than a specific value.

### Two-Tailed Test :
Two-tailed tests test for the possibility of an effect in two directions—positive and negative.

**Two-tailed test: Alternative hypothesis contains the ≠ sign**

Eg:

Suppose a standard fertilizer has been shown to cause a species of plants to grow by an average of 10 inches. However, one botanist believes a new fertilizer causes this species of plants to grow by an average amount different than 10 inches.

   - H0 (Null Hypothesis): μ = 10 inches
   - HA (Alternative Hypothesis): μ ≠ 10 inches

**Note** : This is an example of a two-tailed hypothesis test because the alternative hypothesis contains the not equal “≠” sign. The botanist believes that the new fertilizer will influence plant growth, but doesn’t specify whether it will cause average growth to increase or decrease.

### One Sample Z-Test :
##### Conditions for Z-test :
   - It's performed when population Standard Deviation is given.
   - The minimum Sample size should be 30.
   
Q1. In the population, the average IQ is 100 with a Standard Deviation 15. Researchers want to test a new medication to see if there is a positive or a negative effect on intelligence or no effect at all. A sample of 30 participant who have taken the medication has a mean of 140 and CI is 95%. Did the medication affect the intelligence? 

==> 
   - Null Hypothesis H0       : μ = 100
   - Alternate Hypothesis H1  : μ ≠ 100
   - Significance Value(α) = 0.05
   - Decision : We go ahead with 2 Tailed test as Alternate Hypothesis contains ≠.
   - Z(α/2) = Z(0.05/2) = z(0.025) = +/- 1.96
   - SD (σ) = 15
   - n = 30
   
   **Upper Limit =  100 + 1.96*(15/√30) = 105.36
     lower limit = 100 - 1.96*(15/√30) = 94.63**
   
Here the null hypothesis is rejected as 140 is outside the CI and it can be observed that the medication is increasing the intelligence.

##### Actual way of doing  :
Q2. In the population, the average IQ is 100 with a Standard Deviation 15. Researchers want to test a new medication to see if there is a positive or a negative effect on intelligence or no effect at all. A sample of 30 participant who have taken the medication has a mean of 140 and CI is 95%. Did the medication affect the intelligence? 

==> 
   - Null Hypothesis H0       : μ = 100
   - Alternate Hypothesis H1  : μ ≠ 100
   - Significance Value(α) = 0.05
   - Decision : We go ahead with 2 Tailed test as Alternate Hypothesis contains ≠.
   - Z(α/2) = Z(0.05/2) = z(0.025) = +/- 1.96
   - SD (σ) = 15
   - n = 30
   
**Z = (x - μ)/(σ/√n)**

**Z = (140-100)/(15/√30)**

**Z = 14.60**
   
Here the null hypothesis is rejected as 14.60 is outside the CI +/- 1.96 and it can be observed that the medication is increasing the intelligence.

### One Sample t-Test :
##### Conditions for t-test :
   - It's performed when Sample Standard Deviation is given.
   - The minimum Sample size should be 30.
   
Q1. In the population, the average IQ is 100 . Researchers want to test a new medication to see if there is a positive or a negative effect on intelligence or no effect at all. A sample of 30 participant who have taken the medication has a mean of 140, with a Standard Deviation 15 and CI is 95%. Did the medication affect the intelligence? 

==> 
   - Null Hypothesis H0       : μ = 100
   - Alternate Hypothesis H1  : μ ≠ 100
   - Significance Value(α) = 0.05
   - Decision : We go ahead with 2 Tailed test as Alternate Hypothesis contains ≠.
   - Degree of Freedom = 30-1 = 29 
   - t(α/2) = t(0.05/2) = t(0.025) = +/- 2.045
   - SD (s) = 20
   - n = 30m
   
**t = (x - μ)/(s/√n)**

**t = (140-100)/(20/√30)**

**t = 10.95**
   
Here the null hypothesis is rejected as 10.95 is outside the CI +/- 2.045 range and it can be observed that the medication is increasing the intelligence.

### Chi-Square (Χ²) Tests :
chi-square test is a statistical test for categorical data. It is used to determine whether your data are significantly different from what you expected.

chi-square test are among the most common nonparametric tests. Nonparametric tests are used for data that don't follow the assumptions of parametric tests, especially the assumption of a normal distribution.

Eg. In the 2000 Indian Census, the age of the population in a small town were found to be following :

| Less than 18 | 18 to 35 | Greater than 35 |
|---|---|---|
| 20% | 30% | 50% |

In 2010, ages of 500 people were sampled. Below are the results :

| Less than 18 | 18 to 35 | Greater than 35 |
|---|---|---|
|121|288|91|

Using Significance Value(α) = 0.05, would you conclude the population distribution of the ages has changed in the last 10 years?

===>
   - Expected {Population} :
   
   | Less than 18 | 18 to 35 | Greater than 35 |
   |---|---|---|
   | 20% | 30% | 50% |

   - Observed {Sample n = 500} :
   
   | Less than 18 | 18 to 35 | Greater than 35 |
   |---|---|---|
   |121|288|91|   

   - Expected data based on Population :
   
   || Less than 18 | 18 to 35 | Greater than 35 |
   |---|---|---|---|
   | Observed Data |121|288|91|
   |Expected Data |500x0.2|500x0.3|500x0.5|

   || Less than 18 | 18 to 35 | Greater than 35 |
   |---|---|---|---|
   | Observed Data |121|288|91|
   |Expected Data |100|150|25|
   
   1. $H_0$ = The Data meets the distribution of 2000.
   2. $H_1$ = The Data does not meet the distribution of 2000.
   3. α = 0.05
   4. Degree of Freedom df = no. of categories -1 = 3-1 = 2
   5. Decision : 2 Tailed Test
   6. Check the Chi-Square value for df=2 and α = 0.05 ==> Chi-square($x^2$): 5.991

   
   
   7. Calculate Test Statistics :
      $x^2$ = Σ $(f_0 - f_e)^2$/$f_e$
      
      $x^2$ = $(121-100)^2$/100 + $(288-150)^2$/150 + $(91-250)^2$/250
      
      $x^2$ = 232.94
      
Here, $x^2$ is greater than 5.991 hence we reject the Null Hypothesis.

##### If P-value is greater than Significance Value, indicates strong evidence for the null hypothesis. This means we retain the null hypothesis and reject the alternative hypothesis.
##### If P-value is smaller than Significance Value, we reject the null hypothesis, and accept the alternative hypothesis.

## Central Limit Theorem :
If you take N samples from a population of equal sample size(s) then it will follow 3 things :

    Part 1: If we take N samples from a population P and plot their means, it will follow a Normal Distribution.
    Part 2: Average of the population will be equal to mean of all the sample means.
    Part 3: Standard Deviation of the sample mean distribution is equal to the Population Standard Deviation/sqrt(Number of Data points in each sample(s))
