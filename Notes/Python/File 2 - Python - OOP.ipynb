### Creating a fraction data type

```python
class fraction:
    
    def __init__(self,x,y): ## Parametarized Constructor expects parameters
        self.numerator = x
        self.denominator = y
        
    def __str__(self): ## __MagicMethod__
        return '{}/{}'.format(self.numerator,self.denominator)
    
    def __add__(self,other): ## __add__ this takes in 2 argument 1 self and 2nd other we can name other anything but it's usually called other.
        new_numerator = (self.numerator*other.denominator)+(other.numerator*self.denominator)
        new_denominator = self.denominator*other.denominator
        
        return '{}/{}'.format(new_numerator,new_denominator)
    
    def __sub__(self,other): 
        new_numerator = (self.numerator*other.denominator)-(other.numerator*self.denominator)
        new_denominator = self.denominator*other.denominator
        
        return '{}/{}'.format(new_numerator,new_denominator)
    
    def __mul__(self,other): 
        new_numerator = (self.numerator*other.numerator)
        new_denominator = self.denominator*other.denominator
        
        return '{}/{}'.format(new_numerator,new_denominator)
    
    def __truediv__(self,other): 
        new_numerator = self.numerator*other.denominator
        new_denominator = self.denominator*other.numerator
        
        return '{}/{}'.format(new_numerator,new_denominator)
    
    def convert_decimal(self): ## user defined method
        return self.numerator/self.denominator
    
```

```python
fr1 = fraction(1,2)
fr2 = fraction(1,2)
```

```python
print(fr1+fr2)
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
```

```python
f = fraction(1,2)
f.convert_decimal()
```

1. A user can create and view 2D co-ordinates.
2. A user can find out the distance between 2 co-ordinates
3. A user can find out the distance of a co-ordinateste from origin.
4. A user can check is a point lies on a given line.
5. A user can find the distance between a given 2D point and a line.

```python
class Point:
    
    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y
        
    def __str__(self):
        return '<{}{}>'.format(x_cod,y_cod)
    
    def eucledian_distance(self,other):
        return ((self.x_cod-other.x_cod)**2 + (self.y_cod-other.y_cod)**2)**0.5
    
    def distance_from_origin(self):
        origin = Point(0,0)
        return (self.eucledian_distance(origin))
    
class line:
    
    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C
        
    def __str__(self):
        return '{}x + {}y + {} = 0'.format(self.A,self.B,self.C)
    
    def point_on_line(line,point):
        
        x = point.x_cod
        y = point.y_cod

        if line.A*x + line.B*y + line.C == 0:
            return 'Points ({},{}) lie on line'.format(x,y)
        
        else:
            return 'Points ({},{}) do not lie on line'.format(x,y)
        
    def shortest_distance(line,point):
        
                
        x = point.x_cod
        y = point.y_cod
            
        return abs(line.A*x + line.B*y + line.C)/(line.A**2 + line.B**2)**0.5
        
```

we see that the values do not increment. That is because anything inside a constructor is a instance variable(object variable) and an instance variable starts fresh every time an object is created. So in instance variable, we keep those values which are bound to change with every instance of an object being created. Eg account no., account holder name, account holder gender etc.**self.variable name == instance variable**

where as in a static variable(Class variable), we store the information that is not bound to change eg : Bank name, ifsc code, address etc. **ClassName.variableName == Static variable**

```python
class Atm:
    
    counter = 1 ## Static Variable
    
    def __init__(self):
        print(id(self))
        self.pin = ''
        self.__balance = 0 
        self.cid = Atm.counter  ## Calling a Static Variable
        Atm.counter = Atm.counter + 1 ## Incrementing
```

## Inheritance :

```python
class user:
    
    def __init__(self):
        self.name = 'Prateek'
        self.gender = 'Male'
        
    def login(self):
        print('Login')
        

class Student(user):    ## class Child_Class_Name(Parent_Class_Name): 
                       ## We do not use __init__ in child as if it has a constructor in child it would not look for one in Parent Class.
                       ## & if it doesnot look for constructor in Parent Class, it'd not be able to use them.
        
    def enroll(self):
        print('enroll into the course')
        
u = user()
s = Student()

## From Child Class : It can access attributes/methods of Parent Class
print(s.name) ## From Parent Class
print(s.gender) ## From Parent Class
s.enroll() ## ## From Child Class
```

### Super Keyword :
* This is a keyword used to access the methods of Parent class if Child and Parent both have a class with same name.
* Attributes of Parent Class cannot be accessed with Super() only Methods.
* Super Keyword is used only inside the class and mostly in child class.

```python
class Phone:
    
    def __init__(self,price,brand,camera):
        self.price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):
        print('Buying a Phone')
        
class SmartPhone(Phone):
    
    def buy(self):
        print('Buying a Smart Phone!!!')
        super().buy()   ## Keyword to access buy method of Parent Class.
```

```python
s.buy()
```

### Inheritance in a Summary :

* A class can inherit from another class.
* Inheritance improves code re-useability.
* Constructor, Attributes and Methods get inherited to the child class.
* The parent has no access to the child class.
* Private properties of Parent Class are not accessible directly in the child class.
* Child Class can override the attributes or methods. This is called Method Overriding.
* super() is an inbuilt function, which is used to invoke the parent class methods and constructor.

```python
## Example 1 :

class Parent:
    
    def __init__(self,num): ## 3
        self.__num = num    ## 4
        
    def get_num(self):      ## 5
        return self.__num
    
class Child(Parent):
    
    def __init__(self,num,val): ## 1
        super().__init__(num)   ## 2
        self.__val = val
        
    def get_val(self):     ## 6
        return self.__val
```

```python
son = Child(100,200)
```

```python
print(son.get_num())
print(son.get_val())
```

##### Diamond Problem :
This problem occurs when there are 2 or more Parent Classes and 2 or more Parent classes have a Method by the same name. Now Java does not allow the concept of 2 parent and 1 child but Python does and the Parent class which is passed 1st to the Child Class would be the one, who's Method would be called and executed.

```python
class Phone:
    
    def __init__(self,price,brand, camera):
        print('Inside Phone Constructor')
        self.__price = price
        self.brand = brand
        self.camera = camera
        
    def buy(self):                  ## Same Method Name
        print('Buying a phone')
        
class Product:
    
    def buy(self):                  ## Same Method Name
        print('Product buy Method')
        
class SmartPhone(Product,Phone):    ## Product Class Method executed as it was the 1st class passed to the child.
    pass
```

```python
s = SmartPhone(20000,'Apple',12)
```

```python
s.buy()
```

# Abstraction :
Abstraction means hidden. 
By abstraction, a Parent Class can apply constraints to the Child Class.

```python
from abc import ABC,abstractmethod ## We need to call abc class and that contains abstract method

class BankApp(ABC):  ## it is inheriting from ABC
    
    def database(self):
        print('Connected to Database')
    
    @abstractmethod
    def security(self): ## Abstract Method defined 
                        ## now if any one who wants to access BankApp class or its methods need to have a security Method in their Program.
        pass
```

```python
class MobileApp(BankApp):
    
    def mobile_login(self):
        print('logging into Mobile')
```

```python
m = MobileApp() ## Error, that we need to have Security method inorder to access BankApp class
```

```python
## After implementing security method, we can access MobileApp and also BankApp.


class MobileApp(BankApp):

    def mobile_login(self):
        print('logging into Mobile')

    def security(self):
        print('No Rahul No Pegassis')
```

```python
m = MobileApp()
```
